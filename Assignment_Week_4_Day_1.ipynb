## 1. How do you choose a machine learning algorithm for a problem statement? 
## Explain Briefly.

In order to select the machine learning model, we need to start with the data. If the data is labelled then, it will be a supervised learning model.If the data is not labelled, it will be a unsupervised learning model.

Moving ahead we can focus on the statement itself, if the statement wants us to get the output in terms of continuous data, it will be a regression problem. If the output is to be divided in to classes, it will be a classification algorithm.
If the model is supposed to be divided into multiple clusters, it will be a clustering algorithm.

To reduce Heteroscedasticity, we can transform the dependent variable using some transformation method like log transform. 
To reduct the Heteroscedasticity in regression problems, we can use weighted regression that penalizes the outliers.
