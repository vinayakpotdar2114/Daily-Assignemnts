## 1. An irregular/uneven matrix, or ragged matrix, is a matrix that has a different number of elements in each row. Ragged matrices are not used in linear algebra, since standard matrix transformations cannot be performed on them, but they are useful as arrays in computing.

## Write a Python program to find the indices of all occurrences of target in the uneven matrix.


image.png



image.png

```python
# Write Your Code here.
def search(matrix, target):
    addr = []
    for x, num in enumerate(matrix[0]):
        if num == target:
            addr.append([0, x])
    for row, row_num in zip(matrix[1:], range(1, len(matrix))):
        for x, num in enumerate(row):
            if num == target:
                addr.append([row_num, x])
    return addr
```

```python
matrix = [[1, 3, 2, 32, 19], [19, 2, 48, 19], [], [9, 35, 4], [3, 19]]
target = 19
search(matrix,target)
```

```python
matrix = [[1, 2, 3, 2], [], [7, 9, 2, 1, 4]]
target = 2
search(matrix,target)
```

## 7. What is tradeoff between bias and variances, and the relationship between them?

Bias- The leaning of the data towards a certain class or value can be described as Bias. This Bias can be induced by the way data has been collected.
These biases can be the cognitive or self priorities that the data collector or the model builder might have induced.

Variance- Variance is the measure of the spread unlike the range or interquartile range. This measure of spread accounts in all the data points.
Variance in the data can be caused due to multiple reasons. The variance can caused due to historical shift of the data. It can also be caused due to the random noise.

Now what is Bias Variance tradeoff-
Bias can cause the model to preict the results in favour of a certain class or value, which will cause errors. Variance can lead to the data being spread out and a more generalized model.

Although both of them are equally important, excess of either one would cause overfitting or underfitting. In order to get the model which balances both, we usually tend to reach a point where both are low and the model is performing optimally .

## 8. How overfitting and underfitting can affect model generalization?



Overfitting- If the model learns on a smaller data set or for longer period on the same data, it tends to replicate the results rather than making appropriate predictions. This type of memorization of the data by the model can be described as overfitting.

**The model is not generalized, it wont appropriately predict general or out of sight data**

Underfitting- The phenomenon where the model learns the data for a smaller period of time or has less data to work with. This causes a model that will have consistent errors and won't be able to predict the results properly. 

**This will give less satisfactory results. The model will be more generalized.**

**Overfitting= 1/Model Generalization**

**Underfitting = Model Generalization**
