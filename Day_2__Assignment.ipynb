## 7. Why we cant do a classification problem using Regression?

In regression algorithms, we are trying to create a predictive model where we need continuous data. Classification algorithms use discrete data. 

Discrete data usage-- If we use discrete data, we can set a certain threshold value so that classification can be done easily.

Continuous data usage -- While using continuous data , the algorithms will find it difficult to map values or even maintain a threshold value. This wont help in classification at all. 

Regression tends to make mappings for a set of variables to a certain value.
For classification problem the regression algorithms wont learn properly, as it will try to map multiple sets of variables to a certain class.

## 8. How do you treat heteroscedasticity in regression? What causes Heteroscedasticity and how to fix it?

Hetroscedastic data is a data where the standard deviation is not constant over the period of time.

This is usually caused by the presence of outliers.

In order to remove hetroscedasticity without loosing much of the data, we can perform transformations on the dependent variable.

We can also remove hetroscedasticity by removing the outliers. This will cause some loss of data, so this can be used only when we have abundant data.
