## 7. What is multicollinearity, and how do you treat it and when is it not a problem?

Multicollinearity is a phenomenon where we can observe high degree of correlation between two or more predictor variables.

Multicollinearity will become a problem when the correlation between two or more predictor variables is really high. This will cause the overestimation of the correlated variables. They will be assigned coefficient values which wont be the clear indication of their contribution to the target variable.

In order to understand the degree of multicollinearity, we use VIF i.e. Variance Inflation Factor.

He we calculate the correlation using R2. If the R2 value is high, the VIF is also high. Value of 5+ would be considered as the high degree of correlation and we can ideally eliminate that particular variable.

If the VIF is lower than 5, we can still work with it. 

P.S. The value of 5 is an ideal value, but the threshold can be changed as per the problem at hand.

## 8. VIF(Variation Inflation Factor),Weight of Evidence & Information Value. Why and when to use?

1)VIF- Variance Inflaction Factor-

This is the measure of strength of correlation between different independent variables. VIF comes into picture usually during regression problems.

We use VIF to measure the correlation strength of between different independent variables. Ideally if this value exceeds more than 5, we are supposed to remove that particular feature from the data for further calculations.

VIF is calculated using R2--
VIF= 1/(1-R^2)

2) Weight of Evidence-

As the name suggests, the WOE suggests the power of the predictor variable in correcly predicting the dependent variable. This is mostly used in cases like binary classification.

This term originated from the credit scoring world where it was primarily used to predict the good and bad customers/users of credit system (credit card).

It is calculated as--

WOE= ln(Distribution of Goods/ Distribution of Bads)

3) Information Value(IV)-

IV is again one of the techniques used to select the most important predictor variables.

IV=SUM(%non-event - % of Events)*WOE
