## 1. What is multicollinearity, and how do you treat it and when is it not a problem?

Multicollinearity is a phenomenon in the data that represents the correlation between multiple independent features. This is a problem when one feature is affecting another feature which in turn affects the dependent vairable in multitudes. 
This is not a problem as long as these variables are not affecting the features of interest for the problem statement. eg. year and age. 

Some features can be either removed or combined in order to remove the presence of multicollinearity from the data.

## 3. What is tradeoff between bias and variances, and the relationship between them?

Bias- The amount of difference in the prediction of the model prediction of target value as compared to the training data.

Variance- It is the amount of dispersion in the data. It tells about the degree of spread in the data. In simple words, how much the data points differ from the mean.

Bias and Variance are kind of inversely proportional and we need to reduce both in the data set. In doing so, we will increase either one of them. The point where we can see that both are reduced to a point after which either will increase again is the optimal value of bias and variance acceptable in the model.


## 4. How overfitting and underfitting can affect model generalization?



Overfitting means the model is learning the input data points and will give the same output if a certain feature is encountered. This means that the model is not generalized or not open for different values of input.

Underfitting means when the model does not learn the data properly and is misclassifying or miscalculating the inputs. This means that the model is too generalized. 
We try to find a balance where the model is generalized to a degree where it is able to predict the output to a fair extent whithout outputting the learnt values.
